## The concreteness of visualization for raw data

````python
fig = plt.figure()
ax = plt.axes()
ax.hist(arr, bins=50)
plt.show()
````

## The concreteness of visualization for functions

````python
eps = 1e-06 # epsilon, a tiny number to avoid infinity
p = np.linspace(0 + eps, 1 - eps, 10000)
log_loss_0 = -np.log(1-p)
log_loss_1 = -np.log(p)
````

````python
fig = plt.figure()
ax = plt.axes()
ax.plot(p, log_loss_0, label='$y_{true}=0$')
ax.plot(p, log_loss_1, label='$y_{true}=1$')
ax.legend()
plt.show()
````

## The concreteness of visualization for formulas

````python
x = np.linspace(-10, 10, 100)
fx = 1 / (1 + np.exp(-x))
fig = plt.figure()
ax = plt.axes()
ax.plot(x, fx)
plt.show()
````

## Two ways to plot

## The most challenging part for matplotlib newbies is its various plotting syntaxes

## There are 2 plotting syntaxes to adapt

- Matlab style.
- Object-oriented style.

## Plotting with matlab style: stateful interface

````python
x = np.linspace(0, np.pi*4, 100)
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.show()
````

## Plotting with object-oriented style: object method interface

````python
fig, axes = plt.subplots(2, 1)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()
````

## We will adapt the object-oriented style.

## The general steps

## The general steps to create plots are:

1. Creating arrays.
2. Generating figure and axes objects(respectively or simultaneously).
3. Calling various methods of axes object to generate main plot.
4. Calling various methods of axes object to add elements.
5. Show or save plots.

## Creating arrays

````python
x = np.linspace(0, np.pi*4, 100)
y = np.sin(x)
````

## Generating figure and axes objects respectively

````python
fig = plt.figure()
axes = plt.axes()
````

## Generating figure and axes objects simultaneously

````python
fig, axes = plt.subplots()
````

## Calling various methods of axes object to generate main plot

````python
fig = plt.figure()
axes = plt.axes()
axes.plot(x, y)
````

## Calling various methods of axes object to add elements

````python
fig = plt.figure()
axes = plt.axes()
axes.plot(x, y)
axes.set_title('A simple plot')
axes.set_xlabel('x')
````

## Show plots

````python
fig = plt.figure()
axes = plt.axes()
axes.plot(x, y)
axes.set_title('A simple plot')
axes.set_xlabel('x')
plt.show()
````

## Techniques for adding elements

## How to add title and axis labels

- `set_title` to add title.
- `set_xlabel` to add x-axis label.
- `set_ylabel` to add y-axis label.

````python
fig = plt.figure()
axes = plt.axes()
axes.imshow(arr)
axes.set_title('A very cute Schnauzer.')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
plt.show()
````

## How to adjust axis ticks

- `set_xticks` to adjust x-axis ticks.
- `set_yticks` to adjust y-axis ticks.

````python
fig = plt.figure()
axes = plt.axes()
axes.imshow(arr)
axes.set_title('A very cute Schnauzer.')
axes.set_xticks([])
axes.set_yticks([])
plt.show()
````

## How to adjust axis tick labels

- `set_xticklabels` to adjust x-axis tick labels.
- `set_yticklabels` to adjust y-axis tick labels.

````python
arr = np.random.normal(size=10000)
outside_one_std = np.sum((arr > 1) | (arr < -1)) / arr.size
within_one_std = np.sum((-1 <= arr) & (arr <= 1)) / arr.size
y = [outside_one_std, within_one_std]
x = [1, 2]
````

````python
fig = plt.figure()
axes = plt.axes()
axes.bar(x, y)
axes.set_xticks([1, 2])
axes.set_yticks([0.32,  0.68])
axes.set_xticklabels(['Outside 1 SD', 'Within 1 SD'])
axes.set_yticklabels(['32%', '68%'])
plt.show()
````

## How to adjust axis limits

- `set_xlim` to adjust the upper/lower limits of x-axis.
- `set_ylim` to adjust the upper/lower limits of y-axis.

````python
fig = plt.figure()
axes = plt.axes()
axes.bar(x, y)
axes.set_xticks([1, 2])
axes.set_yticks([0.32,  0.68, 1])
axes.set_xticklabels(['Outside 1 SD', 'Within 1 SD'])
axes.set_yticklabels(['32%', '68%', '100%'])
axes.set_ylim([0, 1])
plt.show()
````

## How to add texts

`text(x, y, 'Some Strings')` to add `'Some Strings'` at (x, y).

````python
fig = plt.figure()
axes = plt.axes()
axes.bar(x, y)
axes.set_xticks([1, 2])
axes.set_yticks([0.32,  0.68, 1])
axes.set_xticklabels(['Outside 1 SD', 'Within 1 SD'])
axes.set_yticklabels(['32%', '68%', '100%'])
axes.set_ylim([0, 1])
axes.text(1 - 0.05, 0.35, '32%')
axes.text(2 - 0.05, 0.70, '68%')
plt.show()
````

## How to add legends

`lenged()` and `label` parameterto add a legend to indicate categories.

````python
x = np.linspace(0, np.pi*4, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
````

````python
fig = plt.figure()
axes = plt.axes()
axes.plot(x, y_sin, label='sin(x)')
axes.plot(x, y_cos, label='cos(x)')
axes.legend()
plt.show()
````

## How to create a subplot grid?

- `subplots(m, n)` to create a `ndarray` with a shape `(m, n)`.
- There are (m, n) axes objects inside the `ndarray`.
- Access axes objects through `[m, n]`.

````python
fig, axes = plt.subplots(2, 3)
print(type(axes))
print(axes.shape)
````

````python
x = np.linspace(0, 4*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, axes = plt.subplots(2, 3)
axes[0, 1].plot(x, y_sin)
axes[1, 2].plot(x, y_cos)
plt.show()
````
